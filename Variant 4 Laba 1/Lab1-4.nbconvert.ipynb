7. Рассмотрим случайный вектор с формой 100х2, представляющей координаты, и найдем расстояния от точки к точке

````python
coordinates = np.random.rand(100, 2)
coordinate_differences = np.diff(coordinates, axis=0)
point_distances = np.sqrt(np.sum(coordinate_differences**2, axis=1))

print(f"Количество точек: {coordinates.shape[0]}")
print(f"Количество расстояний между точками: {len(point_distances)}")
print(f"Все расстояния между последовательными точками:")
print(point_distances)
print(f"\nСтатистика расстояний:")
print(f"Минимальное расстояние: {point_distances.min():.6f}")
print(f"Максимальное расстояние: {point_distances.max():.6f}")
print(f"Среднее расстояние: {point_distances.mean():.6f}")
print(f"Стандартное отклонение: {point_distances.std():.6f}")
````

8. Как конвертировать массив float(32 bit) к массиву целых чисел (integer 32 bit) in place?

````python
def convert_float_to_int_preserving(float_array):
    max_decimals = 0
    for value in float_array:
        if value % 1 != 0:
            decimal_str = f"{value:.6f}".split('.')[1].rstrip('0')
            max_decimals = max(max_decimals, len(decimal_str))
    
    scale = 10 ** min(max_decimals, 100)  
    
    scaled_int_array = (float_array * scale).astype(np.int32)
    
    return scaled_int_array, scale

original_data = np.array([1.7, 2.3, 3.9, 4.1, 5.8], dtype=np.float32)

print("Исходный массив:", original_data)
print("Тип:", original_data.dtype)

simple_int = original_data.astype(np.int32)
print("Простое преобразование:", simple_int)
print("Потеря данных:", original_data - simple_int)

scaled_int, scale_factor = convert_float_to_int_preserving(original_data)
print("Умное преобразование:", scaled_int)

restored_data = scaled_int.astype(np.float32) / scale_factor
print("Восстановленные данные:", restored_data)
print("Точное восстановление:", np.allclose(original_data, restored_data))
````

10. Какой эквивалент enumerate для numpy массивов?

````python
sample_1d = np.array(['a', 'b', 'c', 'd', 'e'])

print("Способ 1 - перебор с индексами:")
for index, value in enumerate(sample_1d):
    print(f"Позиция {index}: значение '{value}'")

print("\nСпособ 2 - многомерное индексирование:")
sample_2d = np.array([[10, 20, 30], [40, 50, 60]])
for coords, val in np.ndenumerate(sample_2d):
    print(f"Координаты {coords}: значение {val}")
````
